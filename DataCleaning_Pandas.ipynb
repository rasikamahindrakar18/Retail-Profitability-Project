# âœ… Step 3: Fixing the Data Type

```python
# Remove commas and convert to float
df['sales'] = df['sales'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)
df['ship_date'] = pd.to_datetime(df['ship_date'], dayfirst=True)

```

# âœ… Step 5: Add Profitability Metrics
## Letâ€™s create a new column called Profit Margin %:

```python
df['Profit Margin %'] = (df['profit'] / df['sales']) * 100
```

## check which columns have negative profit:

```python
loss_data = df[df['profit'] < 0]
print(loss_data[['product_name', 'profit']].head())
```

# âœ… Step 6: Analyze with Grouped Stats
## ðŸ”¹ Category-wise Profit

```python
print(df.groupby('category')['profit'].sum())
```

## ðŸ”¹ Sub-category Profit Margin

```python
subcat_profit = df.groupby('sub_category').agg({
    'sales': 'sum',
    'profit': 'sum'
})
subcat_profit['Profit Margin %'] = (subcat_profit['profit'] / subcat_profit['sales']) * 100
print(subcat_profit)
```
